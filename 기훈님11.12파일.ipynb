```python
import time
```

```python
lat = []
lng = []
for i in range(10):
    part_df = df[300*i:300*(i+1)]
    # ~~~geocode 함수로 주소지를 받아서 위도/경도 값을 받아 위 lat & lon 리스트에 각각 append해주시는 코드~~
    for name in part_df['소재지지번주소']:        
        #gmaps.reverse_geocode((longitude 값, latitude 값), language="ko") == 경도 & 위도 값으로 주소값 가져오기
        #gmaps.geocode('한글 주소', language="ko") 로 위도/경도, 우편번호까지 알 수 있음
        tmpMap = gmaps.geocode(name) # ex) 서울강남경찰서
        tmpLoc = tmpMap[0].get('geometry') # 배열 형태( [~] )로 들어오기 때문에 [0]으로 호출
        lat.append(tmpLoc['location']['lat']) # dict(tmpLoc)의 데이터는 dict['key값'] 로 value 호출
        lng.append(tmpLoc['location']['lng'])
    time.sleep(10)
    
df['lat'] = lat
df['lng'] = lng

df
```

```python
map = folium.Map(location=[35.1507, 126.9139], zoom_start=11)
for n in df.index:
    folium.CircleMarker ([df['lat'][n], df['lng'][n]], 
                         #radius=df['점수'][n]*0.5, # circle 의 크기를 결정
                         color='#3186cc', fill=True, fill_color='#3186cc').add_to(map)
map
```

```python
df_result = {}
search_result = {}

driver = webdriver.Chrome(executable_path='(driver) chromedriver.exe') 
# driver = webdriver.PhantomJS('(driver) phantomjs.exe')

translate_url = 'https://section.blog.naver.com/Search/Post.nhn?pageNo=1&rangeType=ALL&orderBy=sim&keyword='
driver.get(translate_url) 
print(driver.current_url)

 
for key in df_restrant.values(): # 상위 100번째 빈도수에 해당하는 단어까지 담겨있는 dict
    try:
        driver.find_element_by_class_name('textbox').send_keys('광주광역시동구 ' + key) # 여기에 key 대신 다른 값이 들어갔었어요!
        driver.find_element_by_class_name('button').click()
        time.sleep(2)

        web = BeautifulSoup(driver.page_source, 'html.parser')
        result_num = web.find('div', {'class':'search_information'}).find('em', {'class':'search_number'}).get_text()[:-1]
        search_result[key] = result_num

        driver.find_element_by_class_name('textbox').clear()
        time.sleep(2) # 잠깐씩 쉬어주면서 진행합니다.
        
    except NameError:
        print('*** 에러가 발생했습니다 : {}'.format(key))
    except AttributeError:
        print('*** 에러가 발생했습니다 : {}'.format(key))
        
print('전체 번역이 끝났습니다!')

driver.close()
driver.quit()
```
