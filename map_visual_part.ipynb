# 동구음식점 연도별 지도 시각화부분

```python
import numpy as np
import pandas as pd
import seaborn as sns
import folium
import json
```

```python
all_df = pd.read_csv('위도경도채움_주소채움_광주광역시_동구_음식점현황_20200918.csv', encoding='UTF-8')
```

```python
geo_path = 'gwangju_geo_a.json'

geo_str = json.load( open('gwangju_geo_a.json', 'r', encoding='cp949') )
```

```python
print(len(all_df[all_df["영업자시작일"].str.startswith('197')])) #1970년대 식당 9개
print(len(all_df[all_df["영업자시작일"].str.startswith('198')])) #1980년대 식당 14개
print(len(all_df[all_df["영업자시작일"].str.startswith('199')])) # 1990년대 식당 54개
print(len(all_df[all_df["영업자시작일"].str.startswith('200')])) # 2000년대 식당 252개
print(len(all_df[all_df["영업자시작일"].str.startswith('201')])) # 2010년대 식당 2127개
print(len(all_df[all_df["영업자시작일"].str.startswith('202')])) # 2020년대 식당 458개
```

## 빨강색 계열1970,1980,1990,2000 <--> 2010, 2020 파란색계열

```python
map = folium.Map(location=[35.1507, 126.9139], zoom_start=11)  #tiles ='Stamen Terrain')

map.choropleth(geo_data = geo_str,
               fill_color = '#fafad4', 
               key_on = 'feature.id')


for n in all_df.index :
    sub_lat =  all_df.loc[n,'lat']
    sub_long = all_df.loc[n,'lng']
   
    #빨강색 계열1970,1980,1990,2000 <--> 2010, 2020 파란색계열
    
    
    if all_df.loc[n, "영업자시작일"].startswith('197') :  #1970년대 식당 9개
        color = 'red'  #빨강색        
    if all_df.loc[n, "영업자시작일"].startswith('198') : # 1980년대 식당 14개
        color = 'red' #빨강색        
    if all_df.loc[n, "영업자시작일"].startswith('199') : # 1990년대 식당 54개
        color = 'red' #분홍색        
    if all_df.loc[n, "영업자시작일"].startswith('200') : # 2000년대 식당 252개
        color = '#f56858'# 분홍색        
    if all_df.loc[n, "영업자시작일"].startswith('201') : # 2010년대 식당 2127개
        color = '#8dcce3' #하늘색   
    if all_df.loc[n, "영업자시작일"].startswith('202') : # 2020년대 식당 458개
        color = '#465dc2' #파란색
        
    #지도에 동그라미로 데이터 찍기    tooltip=title
    folium.CircleMarker([sub_lat,sub_long],color=color, radius = 0.8).add_to(map)

map
```
